```python
import pandas as pd
import numpy as np
```

```python
scdf = pd.DataFrame([[1,50,80,70,'boy'], 
              [2,60,45,50,'boy'],
              [3,98,43,55,'boy'],
              [4,70,69,89,'boy'],
              [5,56,79,60,'girl'],
              [6,60,68,55,'girl'],
              [7,45,70,77,'girl'],
              [8,55,77,76,'girl'],
              [9,25,57,60,'girl'],
              [10,88,40,43,'girl']], 
                    columns=['student_id',
                             'math_score',
                             'english_score',
                             'chinese_score',
                             'sex'])
scdf = scdf.set_index('student_id')
scdf
```

```python
boysc = scdf.loc[scdf.sex == 'boy']
grlsc = scdf.loc[scdf.sex == 'girl']
print(boysc.mean())
print(grlsc.mean())
```

```python
scdf.groupby('sex').mean()
```

```python
scdf['class'] = [1,2,1,2,1,2,1,2,1,2]
scdf
```

```python
scdf.groupby(['sex', 'class']).mean()
#多欄位一分析
```

```python
scdf.groupby(['sex']).agg(['mean', 'std'])
#一欄位多分析
```

```python
scdf.groupby(['sex', 'class']).agg(['mean', 'std'])
#多欄位多分析
```
