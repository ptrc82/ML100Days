```python
import numpy as np
import pandas as pd
```

```python
a = pd.Series(np.random.randn(200), index = pd.date_range('1/1/2020', periods = 200))
print(a)
acs = a.cumsum()
print(acs)
acs.plot()
```

```python
b = pd.DataFrame(np.random.randn(200,3), index = pd.date_range('1/1/2020', periods = 200), columns = ['1','2','3'])
print(b)
bcs = b.cumsum()
print(bcs)
bcs.plot()
```

```python
#cumsum
c = pd.Series(np.random.randn(200), index = pd.date_range('1/1/2020', periods = 200))
print(c)
ccs = c.cumsum()
print(ccs)
c.plot()
```

```python
e = pd.Series(np.random.rand(4), index = ['1','2','3','4'], name = 'pie plot')
print(e)
e.plot.pie()
```
