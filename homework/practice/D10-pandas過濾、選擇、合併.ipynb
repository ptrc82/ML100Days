```python
stk1 = pd.read_csv('STOCK1.csv')
stk2 = pd.read_csv('STOCK2.csv')
print(stk1,'\n', stk2)
```

```python
print(pd.concat([stk1,stk2], axis = 0, join = 'inner'))
print(pd.concat([stk1,stk2], axis = 1, join = 'inner'))
```

```python
print(pd.concat([stk1,stk2], axis = 0, join = 'inner'))
print(pd.concat([stk1,stk2], axis = 0, join = 'outer'))
```

```python
print(pd.merge(stk1, stk2, on = 'date', how = 'inner'))
print(pd.merge(stk1, stk2, on = 'date', how = 'outer'))
print(pd.merge(stk1, stk2, on = 'date', how = 'left'))
print(pd.merge(stk1, stk2, on = 'date', how = 'right'))
```

```python
stk10in = stk1.set_index('date')
stk20in = stk2.set_index('date')
stk10in.join(stk20in, how = 'outer', lsuffix = '_lf', rsuffix = '_rt')
```
