```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import math
import statistics
```

```python
#2.計算離散均勻分配的概率質量分佈 (probability mass function)
# 之所以稱為質量，是因為離散的點 產生 x 軸的點
low = 1
high = 8
a = np.arange(low,high)
print(a)
probs = stats.randint.pmf(a, low, high)
print(probs)
plt.bar(a, probs)
plt.xlabel('x')
plt.ylabel('P(X=x)')
plt.title('pmf of DU(1,7)')
```

```python
cumprobs = stats.randint.cdf(a, low, high)
print(cumprobs)
plt.plot(a, cumprobs)
plt.xlabel('x')
plt.ylabel('P(X <= x)')
plt.title('cdf of DU(1,7)')
```

```python
b = stats.randint.ppf(cumprobs, low, high)
b
```

```python
c = stats.randint.rvs(low, high, size = 20)
print(c)
plt.hist(c, bins = 13)
plt.show()
```

```python
statt = stats.randint.stats(low, high, moments='mvks')
print(statt)
print(type(statt))
print('mean = ', float(statt[0]))
print('variance = ', float(statt[1]))
print('kurtosis = ', float(statt[2]))
print('skew = ', float(statt[3]))
```

```python
p = 0.8
d = np.arange(0,2)
probs = stats.bernoulli.pmf(d,p)
plt.bar(d, probs)
plt.xlabel('x')
plt.ylabel('P(X=x)')
plt.title('Bernoulli(p = .8)')
plt.show()
```

```python
cumpb = stats.bernoulli.cdf(d,p)
plt.plot(d, cumpb)
plt.xlabel('x')
plt.ylabel('P(X=x)')
plt.title('Bernoulli(p = .8)')
plt.show()
```

```python
pp = stats.bernoulli.ppf(cumpb, p)
pp
```

```python
e = stats.bernoulli.rvs(p, size = 20)
print(e)
plt.hist(e)
plt.show()
```

```python
stber = stats.bernoulli.stats(p, moments = 'mvks')
print(stber)
print(type(stber))
print('bernoulli mean:', float(stber[0]))
print('bernoulli variance:', float(stber[1]))
print('bernoulli kurtosis:', float(stber[2]))
print('bernoulli skew:', float(stber[3]))
```

```python
p2 = 0.82
n = 8
f = np.arange(0, 7)
probs = stats.binom.pmf(f, n, p2)
plt.bar(f, probs)
plt.xlabel('x')
plt.ylabel('P(X=x)')
plt.title('binomial(n=6, p =.82)')
plt.show()
```

```python
csp = stats.binom.cdf(f,n,p2)
plt.ylabel('P(X<=x)')
plt.xlabel('x')
plt.title('binomial(n = 6, p =.82)')
plt.plot(f, csp)
plt.show()
```

```python
ppp = stats.binom.ppf(csp, n, p2)
print(ppp)
```

```python
g = stats.binom.rvs(n, p2, size = 20)
print(g)
plt.hist(g)
plt.show
```

```python
stat_bin=stats.binom.stats(n,p,moments='mvks')
print(stat_bin)
print(type(stat_bin))

print("binomial mean=",float(stat_bin[0]))
print("binomial variance=",float(stat_bin[1]))
print("binomial kurtosis=",float(stat_bin[2]))
print("binomial skew=",float(stat_bin[3]))
```
