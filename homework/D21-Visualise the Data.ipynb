```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.arange(0,820,41)
b = np.sin(a*np.pi/180)
plt.plot(a,b)
plt.show()
```

```python
c = np.arange(0,180)
d = np.sin(c*np.pi/180)
print(plt.plot(c,d))
```

```python
plt.plot(c,d)
plt.xlim(10, 160)
plt.ylim(-1.5, 0.7)
plt.xlabel('axis-c')
plt.ylabel('axis-d')
plt.title('Title')
plt.show()
```
